# Introduction
In this notebook, the basics of connecting to a PostgreSQL database using Python's psycopg2 package will be shown.

Import psycopg2 package:

```python
import psycopg2 as pg2
```

Connect to a PostgreSQL database:

```python
#secret = ''
```

```python
# Connect to the dvdrental database
conn = pg2.connect(database = 'dvdrental', user = 'postgres', password = secret)
```

```python
cur = conn.cursor()
```

Retrieve everything from the payment table of the dvdrental database:

```python
cur.execute('SELECT * FROM payment')
```

```python
# Retrieve the first 5 rows of payment table
data = cur.fetchmany(5)
```

```python
data
```

Notice above return a list of tuples and tuple-unpacking can be used for retrieving information.
